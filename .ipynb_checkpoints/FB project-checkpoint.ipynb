```python
import pandas as pd
df = pd.read_csv("sample.csv", encoding = "ISO-8859-1");
df
#gets a dataframe from the spreadsheet 
```

```python
droppedDf = df.drop(columns=['Date (Responses to recipient forms)', 'Recipient', 'Origin (Responses to recipient forms)', 
                 'Score (Responses to recipient forms)', 'Primary key', 'Age', 'Sex', 'Time Providing Care', 'Race:', 
                'Use of Video Conferencing', 'Other (Connection Technology Use)', 'Consent'])
#df(["Consent"]==1)
#df.dropna(inplace = True) #remove null values
droppedDf


```

```python
coding = {
    'Agree': 1,
    'StronglyAgree': 2,
    'Strongly Agree': 2,
    'StronglyDisagree': -2,
    'Strongly Disagree': -2,
    'Disagree': -1,
    'Yes': 1,
    'No': -1,
    'Rarely':-1,
    'Never':-2,
    'Sometimes':1,
    'Always':2,
    '0':0,
    '1':1
}
def f(s):
    l = []
    for b in s:
        #print(b)
        #print(type(b))
        if (b in coding):
            l.append(coding[b])
        else:
            l.append(b)
    return pd.Series(l)
    #print('\n')
```

```python
ndroppedDf = droppedDf.apply(f, axis = 1);
ndroppedDf
```
